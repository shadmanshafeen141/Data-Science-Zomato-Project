```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

dataframe = pd.read_csv("/kaggle/input/datasets/shadmanshafeen141/zomato-dataset/Zomato-data-.csv")

dataframe['name'][38]
dataframe.head()
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print('Restaurant(s) with the maximum votes:')
print(restaurant_with_max_votes)
print(max_votes)
```

## Bar Plot

```python
x_values = dataframe['listed_in(type)'].unique()
x_values
```

```python
counts = dataframe['listed_in(type)'].value_counts()
print(counts)
y_values = [7 23 8 110]
for i in x_values: 
    y_values.append(counts[i])
```

```python
# fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
# sales = [400, 350, 300, 450]
import matplotlib.pyplot as plt
plt.bar(x_values, y_values)
plt.show()
```
